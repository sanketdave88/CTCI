- Insert

    - at_end
    - at_position
- Delete
    - at_beginning
    - at_end
    - at_position

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        # iterate the linkedlist and print the statements endl with ->
        current = self.head
        while current:
            print(current.data, end = ' -> ')
            current = current.next
        print("None")

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        # iterate through linkedlist
        # when next = null
        # put next = new node
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def insert_at_position(self,data,position):
        # deal with base case if position is 0
        if position == 0:
            self.insert_at_beginning(data)

        current = self.head
        new_node = Node(data)

        for _ in range(position-1):
            if current is None:
                raise(IndexError,"Position out of range")
            current = current.next
            
        new_node.next = current.next
        current.next = new_node

    def delete_at_beginning(self):
        if not self.head:
            raise Exception("List is empty")
        self.head = self.head.next

    def delete_at_end(self):
        if not self.head:
            raise Exception("list is empty")
        
        if self.head.next == None:
            self.head = None
            return

        #iterate until second-last node
        current_node = self.head
        while current_node:
            if current_node.next.next == None:
                current_node.next = None
            current_node = current_node.next

    def delete_at_position(self,position):
        current_node = self.head
        if not current_node:
            raise Exception("List is empty")    

        for _ in range(position):
            if _ == position-1:
                current_node.next = current_node.next.next

            current_node = current_node.next
    
```

```python
linked_list = LinkedList()

linked_list.insert_at_end(5)
linked_list.insert_at_end(10)
linked_list.insert_at_end(15)
linked_list.insert_at_end(20)
linked_list.insert_at_end(25)
linked_list.display()

linked_list.insert_at_position(6,1)
linked_list.display()

linked_list.delete_at_position(2)

linked_list.display()
```
